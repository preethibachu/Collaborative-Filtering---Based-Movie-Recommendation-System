```python
new_name = name.drop(columns=['birthYear', 'deathYear'])
new_name.head()
```

```python
#changing the primary profession columns to indiviual columns and making thr row entries boolean
```

```python
new_name = new_name[~new_name['primaryProfession'].isnull()] 
```

```python
# Split the primaryProfession column into a list of professions
new_name['primaryProfession'] = new_name['primaryProfession'].str.split(',')

# Create a list of unique professions
unique_professions = set(profession for professions in new_name['primaryProfession'] for profession in professions)

# Create a new DataFrame with columns for each unique profession
profession_df = pd.DataFrame()

# Populate the new DataFrame with 1s and 0s
for profession in unique_professions:
    profession_df[profession] = new_name['primaryProfession'].apply(lambda x: 1 if profession in x else 0)

# Concatenate the final DataFrame with the original DataFrame and keep only the relevant columns
result_df = pd.concat([new_name[['nconst', 'primaryName', 'knownForTitles']], profession_df], axis=1)
```

```python
final_merge = final_merge.dropna(subset=['genres', 'averageRating', 'numVotes'])
```

```python
# User preferences
preferred_genres = input("Enter your preferred genres (comma-separated): ").split(',')
min_rating = float(input("Enter the minimum rating you prefer: "))
min_votes = int(input("Enter the minimum number of votes you prefer: "))
```

```python
# Filter movies based on user preferences
filtered_movies = final_merge[
    (final_merge['genres'].str.contains('|'.join(preferred_genres), case=False, na=False)) &
    (final_merge['averageRating'] >= min_rating) &
    (final_merge['numVotes'] >= min_votes)
]
```

```python
# Feature selection
features = ['averageRating', 'numVotes']
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler
```

```python
# Normalize the features using Min-Max scaling
scaler = MinMaxScaler()
final_merge[features] = scaler.fit_transform(final_merge[features])
```

```python
# Create a user profile based on their preferences
user_profile = pd.DataFrame({
    'averageRating': [min_rating],
    'numVotes': [min_votes]
})
```

```python
# Calculate cosine similarity between user profile and all movies
cosine_similarities = cosine_similarity(user_profile[features], final_merge[features])
```

```python
# Add cosine similarities to the filtered_movies DataFrame
final_merge['similarity'] = cosine_similarities[0]
```

```python
# Sort by similarity in descending order
recommended_movies = final_merge.sort_values(by='similarity', ascending=False)
```

```python
# Print the top N recommended movies
N = 10
print(f"Top {N} Recommended Movies:")
for index, row in recommended_movies.head(N).iterrows():
    print(f"{row['title']} ({row['startYear']}), Rating: {row['averageRating']} (Votes: {row['numVotes']})")
```
